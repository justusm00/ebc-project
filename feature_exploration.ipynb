```python
import pandas as pd
```

```python
df_meteo = pd.read_csv("data/meteo_data_preprocessed.csv")
# load already gap fille data
df_fluxes = pd.read_csv("data/flux_data_preprocessed_gapfilled.csv")
```

```python
df = df_meteo.merge(df_fluxes, how="outer", on=["date", "year", "month", "day", "30min", "location"])
```

```python
# select subset where either H_orig or LE_orig are NaN
mask = df[["H_orig", "LE_orig"]].isna().any(axis=1)
subset_df = df[mask]
```

```python
def nan_percentage(df):
    """
    This function takes a DataFrame as input and returns a new DataFrame
    containing the percentage of NaN values in each column.
    
    Parameters:
    df (pd.DataFrame): The input DataFrame.
    
    Returns:
    pd.DataFrame: A new DataFrame with columns 'Column' and 'NaN_Percentage'.
    """
    nan_percent = df.isna().mean() * 100
    result_df = pd.DataFrame({'Column': nan_percent.index, 'NaN_Percentage': nan_percent.values})
    return result_df
```

```python
df_nans = nan_percentage(subset_df)
df_nans
```
